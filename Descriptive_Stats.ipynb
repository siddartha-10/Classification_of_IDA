```python
# Code 1: Anemia Status
code_mapping_anemia = {'Non-Anemic': 0, 'Anemic': 1}
df['Anemia_code'] = df['Anemia_status'].map(code_mapping_anemia)

proportion_anemic = df['Anemia_code'].mean()
total_samples = len(df)
se_anemic = stats.sem(df['Anemia_code'])
ci_anemic = stats.norm.interval(0.95, loc=proportion_anemic, scale=se_anemic)
ci_anemic_percentage = (ci_anemic[0] * 100, ci_anemic[1] * 100)

# Code 2: ID Status
code_mapping_id = {'Non-ID': 0, 'ID': 1}
df['ID_code'] = df['ID_status'].map(code_mapping_id)

proportion_id = df['ID_code'].mean()
se_id = stats.sem(df['ID_code'])
ci_id = stats.norm.interval(0.95, loc=proportion_id, scale=se_id)
ci_id_percentage = (ci_id[0] * 100, ci_id[1] * 100)

# Code 3: IDA Status
proportion_ida = df['IDA_Status'].mean()
se_ida = stats.sem(df['IDA_Status'])
ci_ida = stats.norm.interval(0.95, loc=proportion_ida, scale=se_ida)
ci_ida_percentage = (ci_ida[0] * 100, ci_ida[1] * 100)

# Creating a summary DataFrame
summary_data = {
    'Prevalence (%)': [proportion_anemic * 100, proportion_id * 100, proportion_ida * 100],
    'Standard Error': [se_anemic, se_id, se_ida],
    '95% CI Lower (%)': [ci_anemic_percentage[0], ci_id_percentage[0], ci_ida_percentage[0]],
    '95% CI Upper (%)': [ci_anemic_percentage[1], ci_id_percentage[1], ci_ida_percentage[1]],
    'Sample Size': [total_samples, total_samples, total_samples]
}

summary_df = pd.DataFrame(summary_data, index=['Anemia', 'ID', 'IDA'])

# Displaying the summary table
print(summary_df)

# Stratify by gender
male_df = df[df['RIAGENDR'] == 1]
female_df = df[df['RIAGENDR'] == 2]

# Anemia prevalence by gender
male_anemia_proportion = male_df['Anemia_code'].mean() * 100
female_anemia_proportion = female_df['Anemia_code'].mean() * 100
male_anemia_se = stats.sem(male_df['Anemia_code'])
female_anemia_se = stats.sem(female_df['Anemia_code'])
male_anemia_ci = stats.norm.interval(0.95, loc=male_anemia_proportion, scale=male_anemia_se)
female_anemia_ci = stats.norm.interval(0.95, loc=female_anemia_proportion, scale=female_anemia_se)

print("\nAnemia Prevalence by Gender:")
print(f"Male: {male_anemia_proportion:.2f}% with CI: ({male_anemia_ci[0]:.2f}%, {male_anemia_ci[1]:.2f}%) and SE: {male_anemia_se:.4f}")
print(f"Female: {female_anemia_proportion:.2f}% with CI: ({female_anemia_ci[0]:.2f}%, {female_anemia_ci[1]:.2f}%) and SE: {female_anemia_se:.4f}")

# ID prevalence by gender
male_id_proportion = male_df['ID_code'].mean() * 100
female_id_proportion = female_df['ID_code'].mean() * 100
male_id_se = stats.sem(male_df['ID_code'])
female_id_se = stats.sem(female_df['ID_code'])
male_id_ci = stats.norm.interval(0.95, loc=male_id_proportion, scale=male_id_se)
female_id_ci = stats.norm.interval(0.95, loc=female_id_proportion, scale=female_id_se)

print("\nID Prevalence by Gender:")
print(f"Male: {male_id_proportion:.2f}% with CI: ({male_id_ci[0]:.2f}%, {male_id_ci[1]:.2f}%) and SE: {male_id_se:.4f}")
print(f"Female: {female_id_proportion:.2f}% with CI: ({female_id_ci[0]:.2f}%, {female_id_ci[1]:.2f}%) and SE: {female_id_se:.4f}")

# IDA prevalence by gender
male_ida_proportion = male_df['IDA_Status'].mean() * 100
female_ida_proportion = female_df['IDA_Status'].mean() * 100
male_ida_se = stats.sem(male_df['IDA_Status'])
female_ida_se = stats.sem(female_df['IDA_Status'])
male_ida_ci = stats.norm.interval(0.95, loc=male_ida_proportion, scale=male_ida_se)
female_ida_ci = stats.norm.interval(0.95, loc=female_ida_proportion, scale=female_ida_se)

print("\nIDA Prevalence by Gender:")
print(f"Male: {male_ida_proportion:.2f}% with CI: ({male_ida_ci[0]:.2f}%, {male_ida_ci[1]:.2f}%) and SE: {male_ida_se:.4f}")
print(f"Female: {female_ida_proportion:.2f}% with CI: ({female_ida_ci[0]:.2f}%, {female_ida_ci[1]:.2f}%) and SE: {female_ida_se:.4f}")
```

```python
# Code 1: Anemia Status
code_mapping_anemia = {'Non-Anemic': 0, 'Anemic': 1}
df['Anemia_code'] = df['Anemia_status'].map(code_mapping_anemia)

# Code 2: ID Status
code_mapping_id = {'Non-ID': 0, 'ID': 1}
df['ID_code'] = df['ID_status'].map(code_mapping_id)

# Code 3: IDA Status (if already coded as 1 and 0, no need to map again)
df['IDA_code'] = df['IDA_Status']

# Create a summary DataFrame for all age groups
age_groups_summary = []

# Calculate overall prevalence for reference
total_proportion_anemic = df['Anemia_code'].mean()
total_se_anemic = stats.sem(df['Anemia_code'])
total_ci_anemic = stats.norm.interval(0.95, loc=total_proportion_anemic, scale=total_se_anemic)

total_proportion_id = df['ID_code'].mean()
total_se_id = stats.sem(df['ID_code'])
total_ci_id = stats.norm.interval(0.95, loc=total_proportion_id, scale=total_se_id)

total_proportion_ida = df['IDA_code'].mean()
total_se_ida = stats.sem(df['IDA_code'])
total_ci_ida = stats.norm.interval(0.95, loc=total_proportion_ida, scale=total_se_ida)

# Display the total summary
total_summary_data = {
    'Age Group': 'Overall',
    'Prevalence Anemia (%)': total_proportion_anemic * 100,
    'SE Anemia': total_se_anemic,
    '95% CI Lower Anemia (%)': total_ci_anemic[0] * 100,
    '95% CI Upper Anemia (%)': total_ci_anemic[1] * 100,
    'Prevalence ID (%)': total_proportion_id * 100,
    'SE ID': total_se_id,
    '95% CI Lower ID (%)': total_ci_id[0] * 100,
    '95% CI Upper ID (%)': total_ci_id[1] * 100,
    'Prevalence IDA (%)': total_proportion_ida * 100,
    'SE IDA': total_se_ida,
    '95% CI Lower IDA (%)': total_ci_ida[0] * 100,
    '95% CI Upper IDA (%)': total_ci_ida[1] * 100,
    'Sample Size': len(df)
}

age_groups_summary.append(total_summary_data)

# Creating age groups
bins = [0, 5, 9, 19, float('inf')]
labels = ['<5 years', '5-9 years', '10-19 years', '>=19 years']
df['Age_Group'] = pd.cut(df['RIDAGEYR'], bins=bins, labels=labels, right=False)

# Calculate and display prevalence for each age group
for group in labels:
    age_group_df = df[df['Age_Group'] == group]
    
    proportion_anemic = age_group_df['Anemia_code'].mean()
    se_anemic = stats.sem(age_group_df['Anemia_code'])
    ci_anemic = stats.norm.interval(0.95, loc=proportion_anemic, scale=se_anemic)
    
    proportion_id = age_group_df['ID_code'].mean()
    se_id = stats.sem(age_group_df['ID_code'])
    ci_id = stats.norm.interval(0.95, loc=proportion_id, scale=se_id)

    proportion_ida = age_group_df['IDA_code'].mean()
    se_ida = stats.sem(age_group_df['IDA_code'])
    ci_ida = stats.norm.interval(0.95, loc=proportion_ida, scale=se_ida)
    
    summary_data = {
        'Age Group': group,
        'Prevalence Anemia (%)': proportion_anemic * 100,
        'SE Anemia': se_anemic,
        '95% CI Lower Anemia (%)': ci_anemic[0] * 100,
        '95% CI Upper Anemia (%)': ci_anemic[1] * 100,
        'Prevalence ID (%)': proportion_id * 100,
        'SE ID': se_id,
        '95% CI Lower ID (%)': ci_id[0] * 100,
        '95% CI Upper ID (%)': ci_id[1] * 100,
        'Prevalence IDA (%)': proportion_ida * 100,
        'SE IDA': se_ida,
        '95% CI Lower IDA (%)': ci_ida[0] * 100,
        '95% CI Upper IDA (%)': ci_ida[1] * 100,
        'Sample Size': len(age_group_df)
    }
    
    age_groups_summary.append(summary_data)

# Display the summary table for age groups
print(pd.DataFrame(age_groups_summary))
```

```python
import numpy as np

numerical_columns = ['LBXWBCSI', 'LBXLYPCT', 'LBXMOPCT', 'LBXNEPCT', 'LBXEOPCT', 'LBXBAPCT', 
                      'LBDLYMNO', 'LBDMONO', 'LBDNENO', 'LBDEONO', 'LBDBANO', 'LBXRBCSI', 
                      'LBXHGB', 'LBXHCT', 'LBXMCVSI', 'LBXMC', 'LBXMCHSI', 'LBXRDW', 
                      'LBXPLTSI', 'LBXMPSI', 'LBXFER']

categorical_columns = ['Anemia_code', 'ID_status', 'IDA_Status']

# Create a new DataFrame to store the results
results_list = []

for variable in numerical_columns:
    for category in categorical_columns:
        grouped_data = df.groupby(category)[variable]
        mean_values = grouped_data.mean()
        size_values = grouped_data.size()
        se_values = grouped_data.apply(lambda x: np.std(x, ddof=1) / np.sqrt(len(x)))
        
        # Calculate 95% confidence interval
        ci_values = stats.norm.interval(0.95, loc=mean_values, scale=se_values)
        
        # Append results to the list
        results_list.extend(list(zip([variable] * len(mean_values), [category] * len(mean_values), mean_values.index, mean_values.values, ci_values[0], ci_values[1])))

# Create a DataFrame from the results list
results_df = pd.DataFrame(results_list, columns=['Variable', 'Category', 'Subcategory', 'Mean', 'Lower_CI', 'Upper_CI'])

# Define the new output path to save the Excel file
output_path = r'stats.csv'

# Save the DataFrame to an Excel file
results_df.to_csv(output_path, index=False)
```
